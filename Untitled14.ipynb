```python
# Define image parameters
IMG_WIDTH = 150
IMG_HEIGHT = 150
BATCH_SIZE = 32

# Set up data augmentation and preprocessing for training
train_datagen = ImageDataGenerator(
    rescale=1./255,         # Normalize pixel values
    rotation_range=20,      # Random rotations
    width_shift_range=0.2,  # Random horizontal shifts
    height_shift_range=0.2, # Random vertical shifts
    shear_range=0.2,        # Shear transformations
    zoom_range=0.2,         # Random zooms
    horizontal_flip=True,   # Random horizontal flips
    fill_mode='nearest'     # Strategy for filling pixels exposed by augmentations
)

# Set up data preprocessing for validation/testing (only rescaling)
test_datagen = ImageDataGenerator(rescale=1./255)

# Use flow_from_directory to load images from the specified directories
# and automatically label them based on folder names
train_generator = train_datagen.flow_from_directory(
    TRAIN_DATA_DIR,
    target_size=(IMG_WIDTH, IMG_HEIGHT),
    batch_size=BATCH_SIZE,
    class_mode='categorical' # For classification with multiple classes
)

# Determine which directory to use for validation (val or test)
# Based on the logic from the previous cell
val_test_subdir_name = 'val' if os.path.exists(os.path.join(main_data_dir, 'val')) else 'test'
TEST_DATA_DIR = os.path.join(main_data_dir, val_test_subdir_name)

# If the dataset has a separate validation set, use it here.
# If the dataset only has train/test, you might use a validation_split in the train_datagen
# or use the test_generator for final evaluation after training.
# Based on the previous cell's logic, the dataset seems to have train and either val or test.
# We'll set up the test_generator which can be used for validation during training
# or for final evaluation. If your dataset *specifically* has 'train', 'val', and 'test',
# you would need three generators. Assuming train and val/test structure based on previous output.
validation_generator = test_datagen.flow_from_directory(
    TEST_DATA_DIR, # Use the determined test/val directory for validation during training
    target_size=(IMG_WIDTH, IMG_HEIGHT),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)


# Print the class indices found by the generators (useful for understanding mapping)
print("\nClass indices:")
print(train_generator.class_indices)
print(validation_generator.class_indices) # Should match train_generator

# Get the number of classes
NUM_CLASSES = len(train_generator.class_indices)
print(f"\nNumber of classes detected: {NUM_CLASSES}")

# Check the number of samples found
print(f"Found {train_generator.samples} images belonging to {train_generator.num_classes} classes for training.")
print(f"Found {validation_generator.samples} images belonging to {validation_generator.num_classes} classes for validation.")

# You can now use train_generator and validation_generator for training your model.
# For example: model.fit(train_generator, epochs=..., validation_data=validation_generator)
```

```python
# Build the CNN model
model = models.Sequential([
    # Convolutional layers
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_WIDTH, IMG_HEIGHT, 3)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Flatten layer to transition from convolutional to dense layers
    layers.Flatten(),

    # Dense layers
    layers.Dense(512, activation='relu'),

    # Output layer
    # Use 'softmax' activation for multi-class classification
    layers.Dense(NUM_CLASSES, activation='softmax')
])

# Compile the model
# Use 'categorical_crossentropy' loss for multi-class classification
# Use an optimizer like Adam
# Include 'accuracy' as a metric
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print a summary of the model architecture
model.summary()

print("\nModel built and compiled successfully!")
```

```python
# Define the number of epochs and steps per epoch
# An epoch is one complete pass through the training dataset
# steps_per_epoch = Total number of training samples / Batch size
# validation_steps = Total number of validation samples / Batch size

epochs = 15 # You can adjust this number
# Corrected variable name from batch_size to BATCH_SIZE
steps_per_epoch = train_generator.n // BATCH_SIZE
validation_steps = validation_generator.n // BATCH_SIZE

# Ensure steps are at least 1, especially if dataset is small
if steps_per_epoch == 0:
    steps_per_epoch = 1
if validation_steps == 0:
    validation_steps = 1


# Train the model
# The fit method trains the model using the data generators
print("\nStarting model training...")
history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_steps
)
print("Model training finished.")
```

```python
# Assuming the model training (history) is complete

# Plot training and validation accuracy and loss over epochs
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs) # Use the 'epochs' variable from the training cell

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
# Define the path where the trained model will be saved
MODEL_SAVE_PATH = 'rps_model.h5'

# Save the trained model
print(f"\nSaving the trained model to: {MODEL_SAVE_PATH}")
model.save(MODEL_SAVE_PATH)
print("Model saved successfully!")

# Define the function to determine the winner (assuming basic rock-paper-scissors rules)
def determine_winner(player_move, computer_move):
    # Ensure inputs are lowercase for consistent comparison
    player_move = player_move.lower()
    computer_move = computer_move.lower()

    if player_move == computer_move:
        return 'Draw'
    elif (player_move == 'kertas' and computer_move == 'batu') or \
         (player_move == 'batu' and computer_move == 'gunting') or \
         (player_move == 'gunting' and computer_move == 'kertas'):
        return 'Win'
    else:
        return 'Lose'

# Ensure necessary imports for the game loop
import random
from tensorflow.keras.models import load_model
from IPython.display import display, Javascript, Image as IPyImage
from google.colab.output import eval_js
from base64 import b64decode
import numpy as np
from PIL import Image

# Define target_size using the previously defined IMG_WIDTH and IMG_HEIGHT
target_size = (IMG_WIDTH, IMG_HEIGHT) # Define target_size explicitly

# Define IMAGE_SIZE using target_size
IMAGE_SIZE = target_size

# Assuming idx_to_class was created from the generator in a previous cell
# The class indices are stored in the train_generator.class_indices attribute.
# We need to create a mapping from index to class name.
# Create a dictionary mapping indices to class names
idx_to_class = {v: k for k, v in train_generator.class_indices.items()}

# Derive class names from this mapping
class_names = [idx_to_class[i] for i in sorted(idx_to_class.keys())] # Ensure alphabetical order based on index


def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the display
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename

def preprocess_webcam_image(image_path, target_size=(150, 150)):
    # Load image, convert to RGB, resize, normalize, and add batch dimension
    img = Image.open(image_path).convert('RGB')
    img = img.resize(target_size)
    img_array = np.array(img)
    img_array = img_array / 255.0 # Normalize pixel values to [0, 1]
    img_array = np.expand_dims(img_array, axis=0) # Add batch dimension (1, height, width, channels)
    return img_array

# Load the trained model
loaded_model = load_model(MODEL_SAVE_PATH)
# Corrected the variable name from MODEL_SAVE_SAVE_PATH to MODEL_SAVE_PATH
print(f"\nLoaded model for live prediction from: {MODEL_SAVE_PATH}")
print(f"Model expects input shape: {loaded_model.input_shape}")

# Main loop for live prediction
print("\n--- Rock-Paper-Scissors Game Rules ---")
print("Instructions: Position your hand clearly in front of the webcam.")
print("The model will predict your gesture, and then you'll play against the computer!")
print("Press Enter to capture your gesture. Type 'quit' to exit the game.")

while True:
    user_input = input("\nReady to capture? (Press Enter, or type 'quit'): ")
    if user_input.lower() == 'quit':
        print("Exiting Rock-Paper-Scissors game.")
        break

    try:
        # Capture image from webcam
        filename = take_photo('webcam_capture.jpg')
        print('Image captured. Analyzing gesture...')

        # Display the captured photo
        display(IPyImage(filename))

        # Preprocess the captured image
        # Use the defined IMAGE_SIZE for preprocessing
        processed_image = preprocess_webcam_image(filename, target_size=IMAGE_SIZE)

        # Make prediction
        predictions = loaded_model.predict(processed_image)
        predicted_class_index = np.argmax(predictions, axis=1)[0]
        # Get the predicted class name using the class_names list
        predicted_player_move = class_names[predicted_class_index]
        confidence = np.max(predictions) * 100

        print(f"\nYour Predicted Gesture: **{predicted_player_move.upper()}** (Confidence: {confidence:.2f}%)")

        # --- Game Logic: Computer's Turn and Result ---
        # Ensure computer choices match the class names from the dataset (converted to lowercase)
        computer_choices = [name.lower() for name in class_names]
        computer_move = random.choice(computer_choices)
        print(f"Computer's Move: **{computer_move.upper()}**")

        # Determine and display the game result
        if confidence < 70: # A simple confidence threshold
            print("\nResult: **UNSURE** (Model confidence too low. Please make a clearer gesture and try again.)")
        else:
            # Pass the predicted player move and computer move as lowercase for consistent comparison
            game_result = determine_winner(predicted_player_move.lower(), computer_move.lower())
            print(f"\n--- Game Result: **YOU {game_result.upper()}!** ---")

            if game_result == 'Win':
                print(f"Congratulations! Your {predicted_player_move.upper()} beats the computer's {computer_move.upper()}.")
            elif game_result == 'Lose':
                 # Corrected the print statement to show the computer's move correctly
                print(f"Oops! The computer's {computer_move.upper()} beats your {predicted_player_move.upper()}. Better luck next time!")
            else: # Draw
                print(f"It's a draw! Both you and the computer played {predicted_player_move.upper()}.")

    except Exception as err:
        print(f"\nError during capture or prediction: {err}")
        print("Please ensure you allowed webcam access and your internet connection is stable. Retrying...")

print("\nGame session ended. Thanks for playing!")
```
